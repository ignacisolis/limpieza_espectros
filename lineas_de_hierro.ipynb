# configurar turboespectrum 

```python
def extraer_lineas_hierro(estrella_id, teff, logg, feh, output_file='estrellasyn.par'):
    """crear el archivo de parametros para turbospectrum con las lineas de hierro"""
    config = f"""'LAMBDA_MIN:' '9450.029'
 'LAMBDA_MAX:' '13395.074'
 'LAMBDA_STEP:' '0.01'
 'INTENSITY/FLUX:' 'Flux'
 'COS(THETA)    :' '1.00'
 'ABFIND        :' '.false.' 
 'MODELOPAC:' 'contopac/fe_contopac.dat'
 'RESULTFILE :' 'spectrum_{estrella_id}.spec'
 'METALLICITY:'    '{feh:.2f}'
 'ALPHA/Fe   :'    '0.00'
 'HELIUM     :'    '0.085'
 'R_LIMIT    :'    '20000'
 'MACRO      :'    '0.0'
 'NFILES   :' '1'
 'fe_lines.turbospectrum'"""
    
    with open(output_file, 'w') as f:
        f.write(config)
        print(f"Archivo de parámetros '{output_file}' creado para la estrella {estrella_id} con Teff={teff}, logg={logg}, [Fe/H]={feh}.")
    return output_file
```

```python
ruta=  "/home/nacho/molecfit_test/Turbospectrum2019/estrellasyn.par"
extraer_lineas_hierro('star1', 4681, 1.8, -1.27, output_file=ruta)
```

```python
import os 
import glob
from astropy.io import fits
import matplotlib.pyplot as plt
carpeta = "/home/nacho/molecfit_test/mis_outputs_star1/"
long,flux = interpolar_espectro(carpeta,"*AIR_modificado.par_tac_vcorr.fits",50000,plot=True)
archivos = glob.glob(carpeta + "*AIR_modificado.par_tac_vcorr.fits")

fig, ax = plt.subplots(figsize=(12, 5))  # ← Fuera del loop

for archivo in archivos:
    with fits.open(archivo) as hdul:
        datos = hdul[1].data
        longitud_onda = datos['LAMBDA'] 
        flujo_corregido = datos['FLUX']
        ax.plot(longitud_onda, flujo_corregido, lw=0.5, label=os.path.basename(archivo))

ax.set_xlabel('Longitud de onda (Å)')
ax.set_ylabel('Flujo corregido')
ax.legend(fontsize=6)  # opcional, si hay muchos archivos puede ser útil achicar la leyenda
plt.tight_layout()
plt.show()  # ← También fuera del loop
```
